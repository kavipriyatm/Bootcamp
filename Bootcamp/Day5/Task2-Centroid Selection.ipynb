```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
file_path = "D:\DS-DAY3\CC GENERAL.csv" 
df = pd.read_csv(file_path)
```

```python
X = df[['PURCHASES', 'BALANCE']].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
initializations = ['random', 'k-means++']
plt.figure(figsize=(12, 5))
for i, init in enumerate(initializations, 1):
    kmeans = KMeans(n_clusters=3, init=init, random_state=42, n_init=10)  # Using 3 clusters for comparison
    clusters = kmeans.fit_predict(X_scaled)
    plt.subplot(1, 2, i)
    plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis', marker='o', edgecolors='k', s=100)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
    plt.xlabel('Standardized PURCHASES')
    plt.ylabel('Standardized BALANCE')
    plt.title(f'K-means with {init} initialization')
    plt.legend()

plt.tight_layout()
plt.show()
```
